Project Background

Car insurance companies allocate significant resources toward pricing optimization and risk modeling to estimate the likelihood of customer claims. Because car insurance is mandatory in many countries, the industry represents a large and highly competitive market where accurate prediction of risk is essential for profitability and operational efficiency.

Source: https://www.accenture.com/_acnmedia/pdf-84/accenture-machine-leaning-insurance.pdf

Objective

This project focuses on supporting On the Road Car Insurance in developing a simple and interpretable predictive approach for claim likelihood.

Given limited infrastructure for deploying and maintaining machine learning systems, the objective is to:

Identify the single most predictive feature of insurance claim outcomes using model accuracy as the evaluation metric.

This allows the organization to begin with a lightweight, production-ready baseline model before investing in more complex machine learning systems.



## The dataset

| `outcome` | Whether the client made a claim on their car insurance (response variable): <br><ul><li>`0`: No claim</li><li>`1`: Made a claim</li></ul> |

```python
# Import required modules
import pandas as pd
import numpy as np
from statsmodels.formula.api import logit

# Reading in dataset
df = pd.read_csv("car_insurance.csv")

# Fill missing values with column means
df = df.fillna(df.mean())
```

```python
# ---------------------------------------------
# Preparing for Modeling
# ---------------------------------------------

# Create list of features excluding target and identifier
features = df.drop(columns=["outcome", "id"]).columns

# Store models and accuracy results
models = []
accuracies = []
```

```python
# ---------------------------------------------
# Model Building and Evaluation
# ---------------------------------------------

for col in features:
    
    # Build logistic regression model using a single feature
    model = logit(f"outcome ~ {col}", data=df).fit(disp=0)
    models.append(model)

    # Confusion matrix for predictions
    conf_matrix = model.pred_table()

    # Extract confusion matrix values
    tn = conf_matrix[0,0]
    tp = conf_matrix[1,1]
    fn = conf_matrix[1,0]
    fp = conf_matrix[0,1]
      
    # Calculate accuracy
    accuracy = (tn + tp) / (tn + tp + fn + fp)
    accuracies.append(accuracy)
```

```python
# ---------------------------------------------
# Identify Best Performing Feature
# ---------------------------------------------

best_feature = features[accuracies.index(max(accuracies))]

best_feature_df = pd.DataFrame({
    "best_feature": [best_feature],
    "best_accuracy": [max(accuracies)]
})

print(best_feature_df)
```
